#Conditional Problems:

```python

#1)
apprev = ['A','B','C','D',"F"]
Grade = ['invalid','fail','pass','good','very good','excellent'] 
for a in apprev:
    match a:
        case "invalid":
            print("Invalid grade")
        case "fail":
            print(f"{apprev[-1]} equals to {Grade[1]}")
        case "pass":
            print(f"{apprev[-2]} equals to {Grade[2]}")
        case "good":
            print(f"{apprev[-3]} equals to {Grade[3]}")
        case "very good":
            print(f"{apprev[-4]} equals to {Grade[4]}")
        case "excellent":
            print(f"{apprev[-5]} equals to {Grade[5]}")




#2)
salary = float(input("enter your salary"))

if salary>= 5000:
    print("your bonus will be 2000")
else:
    print("your bomus will be 4000")




#3)
n1 = float(input("Enter the first number: "))
n2 = float(input("Enter the second number: "))
sign = input("Enter the operation you want (+, -, *, /, x): ")

match sign:
    case "-":
        print(f'The result of the operation is {n1 - n2}')
    case "+":
        print(f'The result of the operation is {n1 + n2}')
    case "/":
        if n2 != 0:
            print(f'The result of the operation is {n1 / n2}')
        else:
            print("Error: Division by zero is undefined.")
    case "*" | "x":
        print(f'The result of the operation is {n1 * n2}')
    case _:
        print("Invalid operation entered.")
  
```

# List problems

```python

#1)
Esum = 0
Ecount = 0
Osum = 0
Ocount = 0
lst =[]
x = int(input("Test your luck and enter the number of items you want to be generated"))
for i in range(x):
    lst[i].append(random.randint)

for v in lst:
    if v%2 ==0:
        Esum += v
        Ecount+=1
    else:
        Ocount+=1
        Osum+=v    
print(f'number of even numbers is {Ecount}')
print(f'sum of even numbers is {Esum}')
print(f'number of odd numbers is {Ocount}')
print(f'sum of even numbers is {Osum}')



#2)
lst =[]
x = int(input("enter number of items within a list"))
for i in range(x):
    lst[i].append(random.randint)
print("the list generated is {lst}")
N1= int(input("enter the position you want to swap"))
N2 = int(input("enter the position you want to swap with"))
lst[N1],lst[N2] = lst[N2],lst[N1]
print("list after swapping is {lst}")



#3)
lst =[]
x = int(input("enter number of items within a list that will be randomly generated"))
for i in range(x):
    lst[i].append(random.randint)

num = int(input(f"enter a number to check if it exists in the randomly generated list"))

for i in range(x):
    if lst[i] == num:
        print(f'lucky one, the number exists')
    else:
        print('good luck next time')



#4)
lst =[]
x = int(input("enter number of items within a list that will be randomly generated"))
for i in range(x):
    lst[i].append(random.randint)

num = int(input("enter the number you want to check its occurence times"))
count = 0
for i in range(x):
    if lst[i] == num:
        count+=1
    else:
        print('it doesnt exist LOL')


#5)
lst =[]
x = int(input("enter number of items within a list that will be randomly generated"))
for i in range(x):
    lst[i].append(random.randint)
lst = lst *2



#6)
lst =[]
x = int(input("enter number of items within a list that will be randomly generated"))
for i in range(x):
    lst[i].append(random.randint)
maximum = max(lst)
minimum = min(lst)
print(f'maximum number in the generated list is {maximum}')
print(f'minimum number in the generated list is {minimum}')
```
