```python

def count_word_occurrence(text: str, word_to_count: str):
  """
  Counts the occurrences of a specific word in a given string.

  Args:
    text: The input string.
    word_to_count: The word to count occurrences of.

  Returns:
    The number of times the word appears in the string.
  """
  count = 0
  words = text.split()
  cleaned_word_to_count = ''.join(filter(str.isalnum, word_to_count)).lower()
  for word in words:
    cleaned_word = ''.join(filter(str.isalnum, word)).lower()
    if cleaned_word == cleaned_word_to_count:
      count += 1
  return count

strng = " a computer science portal for portal"
word_to_find = "portal"
occurrence_count = count_word_occurrence(strng, word_to_find)
print(f"The word '{word_to_find}' appears {occurrence_count} times in the string.")
```

```python
def is_palindrome(word: str) -> bool:
  """
  Checks if a given word is a palindrome.

  Args:
    word: The input word.

  Returns:
    True if the word is a palindrome, False otherwise.
  """
  cleaned_word = ''.join(filter(str.isalnum, word)).lower()
  return cleaned_word == cleaned_word[::-1]

word = input("enter a word to check if it is a palindrome: ")
print(f"'{word}' is a palindrome: {is_palindrome(word)}")
```

```python
def sum_positive_numbers(lst: list) -> float:
  """
  Calculates the sum of positive numbers in a list.

  Args:
    lst: The input list of numbers.

  Returns:
    The sum of positive numbers in the list.
  """
  p_sum = 0
  for num in lst:
     if num > 0:
        p_sum += num
  return p_sum

lst = []
N = int(input("enter the length of the list: "))
for i in range(N):
  num = int(input(f"enter the {i+1}th number: "))
  lst.append(num)

print(f'the sum of the positive numbers in the list is {sum_positive_numbers(lst)}')
```
